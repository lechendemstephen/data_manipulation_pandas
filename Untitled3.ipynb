```python
passed = df[(df['math score']>90) & (df['reading score']>90) & (df['writing score']>90)]
```

```python
passed_females = passed[(passed['gender'] == 'female') & (passed['parental level of education'] == "bachelor's degree")]
passed_females
passed_females.to_csv('passed_females.csv')
```

```python
passed_males = passed[(passed['gender'] == 'male') & (passed['parental level of education'] == "bachelor's degree")]
passed_males.to_csv('passed_males.csv')
```

```python
df['Total Score'] = df['math score'] + df['reading score'] + df['writing score']
```

```python
failed_students = df[df['Total Score'] <=150]
failed_students
```
