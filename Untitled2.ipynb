```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

gas = pd.read_csv('gas_prices.csv')
gas.head(5)
```

```python


#germany = plt.plot(gas['Year'], gas['Germany'])
#France = plt.plot(gas['Year'], gas['France'])
lables = ['Australia', 'Canada', 'France', 'Germany', 'Italy', 'Japan']
[plt.plot(gas.Year, gas[country], marker ='.', label=country) for country in gas if country != 'Year']

plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('Price($)')
plt.title('Change in Fuel Price(USD)')
plt.legend()
plt.figure(figsize=(10, 6), dpi=600)
plt.show()
```
